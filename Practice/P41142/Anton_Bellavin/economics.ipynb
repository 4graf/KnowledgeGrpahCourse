```python
import yfinance as yf
import pandas as pd
import datetime
import numpy as np

tickers_spbex_csv = pd.read_csv('spbex.csv', encoding='cp1251', sep=';')
raw_spbex_tickers = [ticker.replace('@', '.') for ticker in tickers_spbex_csv['s_RTS_code'].dropna().unique()]

tickers_mobex_csv = pd.read_csv('mobex.csv', encoding='cp1251', sep=';')
raw_mobex_tickers = [str(ticker).replace('-RM', '') for ticker in tickers_mobex_csv['TRADE_CODE'].dropna().unique()]

raw_tickers = sorted(np.unique(raw_spbex_tickers + raw_mobex_tickers))

tickers = ' '.join([str(ticker) for ticker in raw_tickers])

start_date = datetime.datetime.now()
end_date = datetime.datetime.now()
format_time = '%Y-%m-%d'

start_date_str = start_date.strftime(format_time)
end_date_str = end_date.strftime(format_time)

print("count of tickers = ", len(tickers))
print(tickers)
```

```python
#data = yf.download(tickers, start=start_date_str, end=end_date_str)
data = yf.Tickers(tickers)
```

```python
import sys
import os
import csv

tickers_list = list(data.tickers.items())
header = []
rows = []
data_length = len(tickers_list[:600])
for idx, value in enumerate(tickers_list[:600]):
    ticker = value[1]
    info = ticker.get_info()
    if 'quoteType' in info:
      if idx == 0:
        header = info.keys()
      rows.append([x[1] for x in list(info.items())])

    log_str = "tickers info: " + str(idx + 1) + " of " + str(data_length)
    sys.stdout.write("\r" + log_str)
    sys.stdout.flush()
os.system('cls')

with open('result.csv', 'w', newline='') as out_file:
  tsv_writer = csv.writer(out_file)
  tsv_writer.writerow(header)
  for row in rows:
    tsv_writer.writerow(row)


```
