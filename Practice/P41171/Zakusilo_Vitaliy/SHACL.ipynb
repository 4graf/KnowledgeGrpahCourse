# SHACL

## shapes

```python
base_uri = 'http://www.semanticweb.org/basicec/ontologies/2021/4/animes'
base_path = '/content/drive/MyDrive/AnimeData/'

shapes_file = '''
@prefix ansh: <http://www.semanticweb.org/basicec/ontologies/2021/4/animes_shapes#> .
@prefix an: <http://www.semanticweb.org/basicec/ontologies/2021/4/animes#> .
@prefix owl: <http://www.w3.org/2002/07/owl#> .
@prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> .
@prefix rdfs: <http://www.w3.org/2000/01/rdf-schema#> .
@prefix sh: <http://www.w3.org/ns/shacl#> .
@prefix xsd: <http://www.w3.org/2001/XMLSchema#> .
 
ansh:StudioShape
    a sh:NodeShape ;
    sh:targetClass an:Studio ;
    sh:property [
        sh:path an:name;
        sh:maxCount 1 ;
        sh:minCount 1 ;
        sh:datatype xsd:string ;
    ] ;
.

ansh:GenreShape
    a sh:NodeShape ;
    sh:targetClass an:Genre ;
    sh:property [
        sh:path an:name;
        sh:maxCount 1 ;
        sh:minCount 1 ;
        sh:datatype xsd:string ;
    ] ;
.

ansh:PersonShape
    a sh:NodeShape ;
    sh:targetClass an:Person ;
    sh:property [
        sh:path an:name;
        sh:maxCount 1 ;
        sh:datatype xsd:string ;
    ] ;
    sh:property [
        sh:path an:birthday;
        sh:maxCount 1 ;
        sh:datatype xsd:string ;
    ] ;
.

ansh:AnimeShape
    a sh:NodeShape ;
    sh:targetClass an:Anime ;
    sh:property [
        sh:path an:name;
        sh:maxCount 1 ;
        sh:datatype xsd:string ;
    ] ;
    sh:property [
        sh:path an:hasRating;
        sh:maxCount 1 ;
    ] ;
    sh:property [
        sh:path an:hasKind;
        sh:maxCount 1 ;
    ] ;
.
'''
shapes_file_format = 'turtle'
```

## validation

```python
from pyshacl import validate

conforms, v_graph, v_text = validate(f'{base_path}Anime_r.rdf', shacl_graph=shapes_file,
                                     target_graph_format='rdfxml',
                                     shacl_graph_format=shapes_file_format,
                                     inference='rdfs', debug=True,
                                     serialize_report_graph=True)
print(conforms)
```
