```python
from rdflib import Graph, URIRef, BNode, Literal
import pandas
from rdflib.namespace import RDF
```

# Loading dataset

```python
df = pandas.read_excel('./data/dataset.xlsx', index_col=0)
df[:5]
```

```python
g = Graph()
g.parse("films.owl", format="turtle")
```

# Classes, object property and data property

```python
path = 'http://www.kinopoisk.ru/films-ontology#'

# class
Film_class = URIRef(path + 'Film')
GoodFilm_class = URIRef(path + 'GoodFilm')
BadFilm_class = URIRef(path + 'BadFilm')
Genre_class = URIRef(path + 'Genre')
Staff_class = URIRef(path + 'Staff')
Actor_class = URIRef(path + 'Actor')
Director_class = URIRef(path + 'Director')
Writer_class = URIRef(path + 'Writer')
Adventures_class = URIRef(path + 'Adventures')
Detective_class = URIRef(path + 'Detective')
Horror_class = URIRef(path + 'Horror')
Drama_class = URIRef(path + 'Drama')
Melodrama_class = URIRef(path + 'Melodrama')
Comedy_class = URIRef(path + 'Comedy')
Historical_class = URIRef(path + 'Historical')
Thriller_class = URIRef(path + 'Thriller')
Catroon_class = URIRef(path + 'Cartoon')
Military_class = URIRef(path + 'Military')
Fantasy_class = URIRef(path + 'Fantasy')
Crime_class = URIRef(path + 'Crime')
Biography_class = URIRef(path + 'Biography')
Family_class = URIRef(path + 'Family')

#object property
hasActor_objprop = URIRef(path + 'hasActor')
hasDirector_objprop = URIRef(path + 'hasDirector')
hasWriter_objprop = URIRef(path + 'hasWriter')
hasGenre_objprop = URIRef(path + 'hasGenre')
participated_objprop = URIRef(path + 'participated')

#data property
budget_dataprop = URIRef(path + 'budget')
coutry_dataprop = URIRef(path + 'coutry')
description_dataprop = URIRef(path + 'description')
filmLength_dataprop = URIRef(path + 'filmLength')
hasName_dataprop = URIRef(path + 'hasName')
hasRating_dataprop = URIRef(path + 'hasRating')
namedFilmEn_dataprop = URIRef(path + 'namedFilmEn')
namedFilmRu_dataprop = URIRef(path + 'namedFilmRu')
ratingAgeLimit_dataprop = URIRef(path + 'ratingAgeLimit')
year_dataprop = URIRef(path + 'year')
```

# Function which added films with staff and genres

```python
import re

def add_film(nameFilmEn, nameFilmRu, budget, countries, 
             description, filmLength, rating, ratingAgeLimit, year,
            staff_actor, staff_writer, staff_director, genres):
    film_id = str(re.sub('[^0-9А-Яа-я]', '', str(nameFilmRu)))
    film = (URIRef(path + film_id))
    g.add((film, RDF.type, Film_class))
    g.add((film, budget_dataprop, Literal(budget)))
    g.add((film, description_dataprop, Literal(description)))
    g.add((film, filmLength_dataprop, Literal(filmLength)))
    g.add((film, hasRating_dataprop, Literal(rating)))
    g.add((film, namedFilmEn_dataprop, Literal(nameFilmEn)))
    g.add((film, namedFilmRu_dataprop, Literal(nameFilmRu)))
    g.add((film, ratingAgeLimit_dataprop, Literal(ratingAgeLimit)))
    g.add((film, year_dataprop, Literal(year)))
        
    for country in countries:
        g.add((film, coutry_dataprop, Literal(country)))
    
    for genre in genres.strip('][').split(', '):
        if genre == "'приключения'":
            g.add((film, RDF.type, Adventures_class))
            g.add((film, hasGenre_objprop, Literal(genre.replace("'", ""))))
        elif genre == "'детектив'":
            g.add((film, RDF.type, Detective_class))
            g.add((film, hasGenre_objprop, Literal(genre.replace("'", ""))))
        elif genre == "'ужасы'":
            g.add((film, RDF.type, Horror_class))
            g.add((film, hasGenre_objprop, Literal(genre.replace("'", ""))))
        elif genre == "'драма'":
            g.add((film, RDF.type, Drama_class))
            g.add((film, hasGenre_objprop, Literal(genre.replace("'", ""))))
        elif genre == "'мелодрама'":
            g.add((film, RDF.type, Melodrama_class))
            g.add((film, hasGenre_objprop, Literal(genre.replace("'", ""))))
        elif genre == "'комедия'":
            g.add((film, RDF.type, Comedy_class))
            g.add((film, hasGenre_objprop, Literal(genre.replace("'", ""))))
        elif genre == "'вестерн'":
            g.add((film, RDF.type, Military_class))
            g.add((film, hasGenre_objprop, Literal(genre.replace("'", ""))))
        elif genre == "'фэнтези'":
            g.add((film, RDF.type, Fantasy_class))
            g.add((film, hasGenre_objprop, Literal(genre.replace("'", ""))))
        elif genre == "'мультфильм'":
            g.add((film, RDF.type, Catroon_class))
            g.add((film, hasGenre_objprop, Literal(genre.replace("'", ""))))
        elif genre == "'триллер'":
            g.add((film, RDF.type, Thriller_class))
            g.add((film, hasGenre_objprop, Literal(genre.replace("'", ""))))
        elif genre == "'история'":
            g.add((film, RDF.type, Historical_class))
            g.add((film, hasGenre_objprop, Literal(genre.replace("'", ""))))
        elif genre == "'криминал'":
            g.add((film, RDF.type, Crime_class))
            g.add((film, hasGenre_objprop, Literal(genre.replace("'", ""))))
        elif genre == "'биография'":
            g.add((film, RDF.type, Biography_class))
            g.add((film, hasGenre_objprop, Literal(genre.replace("'", ""))))
        elif genre == "'семейный'":
            g.add((film, RDF.type, Family_class))
            g.add((film, hasGenre_objprop, Literal(genre.replace("'", ""))))
            
    
    staff_actor = staff_actor.strip('][').split(', ')
    staff_writer = staff_writer.strip('][').split(', ')
    staff_director = staff_director.strip('][').split(', ')
    
    for actor in staff_actor:
        actor_id = (URIRef(path + str(actor).replace(" ", "").replace("'", "").replace("\"", "")))
        g.add((actor_id, RDF.type, Actor_class))
        g.add((actor_id, hasName_dataprop, Literal(actor)))
        g.add((actor_id, participated_objprop, film))
        g.add((film, hasActor_objprop, actor_id))
    
    for writer in staff_writer:
        writer_id = (URIRef(path + str(writer).replace(" ", "").replace("'", "").replace("\"", "")))
        g.add((writer_id, RDF.type, Writer_class))
        g.add((writer_id, hasName_dataprop, Literal(writer)))
        g.add((writer_id, participated_objprop, film))
        g.add((film, hasWriter_objprop, writer_id))
                    
    for director in staff_director:
        director_id = (URIRef(path + str(director).replace(" ", "").replace("'", "").replace("\"", "")))
        g.add((director_id, RDF.type, Director_class))
        g.add((director_id, hasName_dataprop, Literal(director)))
        g.add((director_id, participated_objprop, film))
        g.add((film, hasDirector_objprop, director_id))
    return print(film)
```

# Add individuals

```python
for i in range(len(df)):
    add_film(df.iloc[i]['nameEn'], df.iloc[i]['nameRu'], df.iloc[i]['budget'], 
             df.iloc[i]['countries'], df.iloc[i]['description'], df.iloc[i]['filmLength'], 
             df.iloc[i]['rating'], df.iloc[i]['ratingAgeLimits'], df.iloc[i]['year'], 
             df.iloc[i]['actor'], df.iloc[i]['writer'], df.iloc[i]['director'], df.iloc[i]['genres'])  
```

# Save to films_with_individuals.owl

```python
g.serialize(destination = 'films_with_individuals.owl', format = 'turtle')
```
