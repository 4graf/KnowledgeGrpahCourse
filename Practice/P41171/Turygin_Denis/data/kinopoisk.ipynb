```python
import requests
import pandas as pd
import time
import numpy as np
```

```python
headers = {
    'accept': 'application/json',
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36',
    'X-API-KEY': '1b312789-c1ed-4e9b-9803-6372843df882',
}

params = (
    ('append_to_response', ['BUDGET', 'RATING', 'REVIEW']),
)
```

```python
data = []
value_id = 1000

for id in range(300, 100500):
    response_metadata = requests.get('https://kinopoiskapiunofficial.tech/api/v2.1/films/' + str(id),
                            headers=headers, params=params)
    time.sleep(1)
    response_staff = requests.get('https://kinopoiskapiunofficial.tech/api/v1/staff?filmId=' + str(id),
                            headers=headers)
    if (response_metadata.status_code & response_staff.status_code) == 200 :
        response_metadata = response_metadata.json()
        response_staff = response_staff.json()
        results = {
            'nameRu': response_metadata['data']['nameRu'],
            'nameEn': response_metadata['data']['nameEn'],
            'year': response_metadata['data']['year'],
            'filmLength': response_metadata['data']['filmLength'],
            'description': response_metadata['data']['description'],
            'type': response_metadata['data']['type'],
            'ratingMpaa': response_metadata['data']['ratingMpaa'],
            'ratingAgeLimits': response_metadata['data']['ratingAgeLimits'],
            'distributors': response_metadata['data']['distributors'],
            'distributorRelease': response_metadata['data']['distributorRelease'],
            'countries': [i['country'] for i in response_metadata['data']['countries']],
            'genres':  [i['genre'] for i in response_metadata['data']['genres']],
            'seasons': response_metadata['data']['seasons'],
            'rating': response_metadata['rating']['rating'],
            'budget': response_metadata['budget']['budget'],
            'actor': [i['nameEn'] for i in response_staff if i['professionKey'] == 'ACTOR'],
            'writer': [i['nameEn'] for i in response_staff if i['professionKey'] == 'WRITER'],
            'director': [i['nameEn'] for i in response_staff if i['professionKey'] == 'DIRECTOR'],
        }

        if id == value_id:
            value_id += 500
            print('need sleep 1 minute')
            time.sleep(30)
        print('{}: Film "{}" appended'.format(id, results['nameRu']))
        data.append(results)
```

```python
df = pd.DataFrame(data)
df.to_excel("output.xlsx")  
```
